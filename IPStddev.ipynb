# What are spatial domain operations?
The term spatial refers to space. In an image, this space is a 2D plane (xy-plane). So, the spatial domain refers to the image plane itself and methods in spatial domain are based on directly modifying the value of the pixels.

## What is Standard Deviation?

The Standard Deviation is a measure of how spread out numbers are.<br>
Its symbol is Ïƒ (the greek letter sigma).<br>
It is the square root of the variance of the data.<br>
In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range.<br>
